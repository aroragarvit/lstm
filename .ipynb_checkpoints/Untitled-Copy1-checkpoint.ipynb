```python
input_features=Stock_data[Stock_data['Date'] >= '2017-11-09']
input_features=input_features[["Open","High","Low","Close","Volume"]]
print("Shape of input features" ,input_features.shape)

opening_prices=input_features["Open"]
print("Shape of opening Prices" ,opening_prices.shape)

```

```python
# days for test data =  days in last for whiich you want to test
days_for_test_data=100
days_sequence=50
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler2 = MinMaxScaler()

normalised_input_features = scaler.fit_transform(input_features.values)
normalised_opening_prices=scaler2.fit_transform((opening_prices).values.reshape(-1,1))

joblib.dump(scaler,"ethscaler.save")
joblib.dump(scaler2,"ethscaler2.save")

train_normalised_input_features=normalised_input_features[0:input_features.shape[0]-days_for_test_data,:]  # starting ka 30 ka kuch nhi nikalaga 
train_normalised_opening_prices=normalised_opening_prices[0:input_features.shape[0]-days_for_test_data,:]

print("Shape of train normalised input features" ,train_normalised_input_features.shape)
print("Shape of train normalised opening prices " ,train_normalised_opening_prices.shape)

test_normalised_input_features=normalised_input_features[input_features.shape[0]-days_for_test_data-days_sequence:input_features.shape[0],:]  # isma bhi apan log x_train ka 30 le raha ha kyuki starting ka 30 ka kuch nahi nikalaga
test_normalised_opening_prices=normalised_opening_prices[input_features.shape[0]-days_for_test_data-days_sequence:input_features.shape[0],:]

print("Shape of test normalised input features " ,test_normalised_input_features.shape)
print("Shape of test normalised opening prices " ,test_normalised_opening_prices.shape)

test_input_features=input_features.values[input_features.shape[0]-days_for_test_data-days_sequence:input_features.shape[0],:]
print("Shape of test input features" ,test_input_features.shape)

test_opening=opening_prices.values.reshape(-1,1)[input_features.shape[0]-days_for_test_data-days_sequence:input_features.shape[0],:]
print("Shape of test opening prices" ,test_opening.shape)

train_input_features=input_features.values[0:input_features.shape[0]-days_for_test_data,:]
print("Shape of train input features", train_input_features.shape)

train_opening=opening_prices.values.reshape(-1,1)[0:input_features.shape[0]-days_for_test_data,:]
print( "Shape of train opeining prices", train_opening.shape)

# train_normalised_input_features
# train_normalised_opening_prices
# train_input_features
# train_opening

# test_normalised_input_features
# test_normalised_opening_prices
# test_input_features
# test_opening
np.set_printoptions(threshold=np.inf)
np.set_printoptions(suppress=True)
print(normalised_input_features)
```

```python
x_train=[] # these are normalised ones 
y_train=[]
x_train_original=[]
y_train_original=[]
for i in range(days_sequence,train_normalised_input_features.shape[0]):
  x_train.append(train_normalised_input_features[i-days_sequence:i,:])
  y_train.append(train_normalised_opening_prices[i,0])
  
  x_train_original.append(train_input_features[i-days_sequence:i,:])
  y_train_original.append(train_opening[i,0])

x_train_original,y_train_original=np.array(x_train_original),np.array(y_train_original)  
x_train,y_train=np.array(x_train),np.array(y_train)
print("Shapes of training sequences")
print(x_train.shape)
print(y_train.shape)
print(x_train_original.shape)
print(y_train_original.shape)

x_test_normalised=[]
y_test_normalised=[]
x_test=[]
y_test=[]

for i in range(days_sequence,test_normalised_input_features.shape[0]):
  x_test_normalised.append(test_normalised_input_features[i-days_sequence:i,:])
  y_test_normalised.append(test_normalised_opening_prices[i,0])
  x_test.append(test_input_features[i-days_sequence:i,:])
  y_test.append(test_opening[i,0])


x_test,y_test=np.array(x_test),np.array(y_test)
print(x_test.shape)
print(y_test.shape)

x_test_normalised,y_test_normalised=np.array(x_test_normalised),np.array(y_test_normalised)
print(x_test_normalised.shape)
print(y_test_normalised.shape)
```

```python
def create_model():
    model=Sequential()
    model.add(LSTM(units=10,input_shape = (x_train.shape[1],x_train.shape[2]),activation="relu",return_sequences=True))
    model.add(LSTM(units=10,input_shape = (x_train.shape[1],x_train.shape[2]),activation="relu"))
    model.add(Dropout(0.1))
    model.add(Dense(1))
    model.add(BatchNormalization())
    model.compile(optimizer = 'adam', loss = 'mean_squared_error',metrics=[tf.keras.metrics.RootMeanSquaredError()])
    
    return model

model = create_model()
model.summary()
                  
```

```python
model_fit=model.fit(x_train, y_train,validation_split=0.1, epochs = 200, batch_size = 30)

model.summary()
```

```python
plt.plot(model_fit.history['loss'])
plt.plot(model_fit.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
training_opening_predictions = scaler2.inverse_transform(model.predict(x_train))
testing_opening_predictions=scaler2.inverse_transform(model.predict(x_test_normalised))
```
